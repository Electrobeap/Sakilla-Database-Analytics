```python
def fetch_query(query, db_path):
    """
    Executes a SQL query against a database and returns the results as a Pandas DataFrame.

    Args:
      query: The SQL query to execute.
      db_path: The path to the database file.

    Returns:
      A Pandas DataFrame containing the query results.
    """
    try:
        conn = sqlite3.connect(db_path)
        df = pd.read_sql_query(query, conn)
        return df
    except Exception as e:
        print(f"Error executing query: {e}")
        return None

#Top-Rented Movies with rental pattern by demographics
query = """
WITH rental_counts AS (
  SELECT
    f.film_id AS film_id,
    f.title AS title,
    COUNT(r.rental_id) AS rental_count
  FROM film f
  JOIN inventory i ON f.film_id = i.film_id
  JOIN rental r ON i.inventory_id = r.inventory_id
  GROUP BY f.film_id, f.title
  ORDER BY rental_count DESC
),
customer_demographics AS (
  SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.email,
    c.address_id,
    a.address,
    a.district,
    a.city_id,
    ci.city,
    ci.country_id,
    co.country,
    c.store_id,
    c.active,
    c.last_update,
    CAST(STRFTIME('%Y', 'now') AS INTEGER) - CAST(STRFTIME('%Y', c.create_date) AS INTEGER) AS customer_age,
    CASE
      WHEN CAST(STRFTIME('%Y', c.create_date) AS INTEGER) %2 = 0 THEN 'Female'
      ELSE 'Male'
    END AS gender
  FROM customer c
  JOIN address a ON c.address_id = a.address_id
  JOIN city ci ON a.city_id = ci.city_id
  JOIN country co ON ci.country_id = co.country_id
),
rental_demographics AS (
  SELECT
    r.rental_id,
    r.rental_date,
    r.inventory_id,
    r.customer_id,
    cd.first_name,
    cd.last_name,
    cd.email,
    cd.address,
    cd.district,
    cd.city,
    cd.country,
    cd.store_id,
    cd.active,
    cd.customer_age,
    cd.gender
  FROM rental r
  JOIN customer_demographics cd ON r.customer_id = cd.customer_id
),
top_rented_movies AS (
SELECT
  rc.film_id,
  rc.title,
  rc.rental_count,
  rd.customer_id,
  rd.first_name,
  rd.last_name,
  rd.email,
  rd.address,
  rd.district,
  rd.city,
  rd.country,
  rd.store_id,
  rd.active,
  rd.customer_age,
  rd.gender
FROM rental_counts rc
JOIN inventory i ON rc.film_id = i.film_id -- Corrected join condition
JOIN rental r ON i.inventory_id = r.inventory_id
JOIN rental_demographics rd ON r.customer_id = rd.customer_id
)
SELECT
  title,
  rental_count,
  customer_id,
  first_name,
  last_name,
  email,
  address,
  district,
  city,
  country,
  store_id,
  active,
  customer_age,
  gender
FROM top_rented_movies
ORDER BY rental_count DESC, customer_age ASC;
"""

# Replace 'your_database.db' with the actual path to your database
top_rented_movies_with_demographics = fetch_query(query, db_path='/content/drive/MyDrive/Capstone project/sakilla database.db')
print(top_rented_movies_with_demographics)
```

```python
# customer segment based on their rental behaviour
query = """
WITH rental_counts AS (
  SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    COUNT(r.rental_id) AS total_rental,
    MAX(r.rental_date) AS last_rental_date
  FROM customer c
  JOIN rental r ON c.customer_id = r.customer_id
  GROUP BY c.customer_id, c.first_name, c.last_name
),
customer_payment AS (
  SELECT
    p.customer_id,
    SUM(p.amount) AS total_payments
  FROM payment p
  GROUP BY p.customer_id
),
  customer_segment AS (
  SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    c.total_rental,
    c.last_rental_date,
    cp.total_payments,
    CASE
      WHEN cp.total_payments > 200 THEN 'High Value'
      WHEN cp.total_payments BETWEEN 100 AND 200 THEN 'Medium Value'
      ELSE 'Low Value'
    END AS customer_segment
  FROM rental_counts c
  LEFT JOIN customer_payment cp ON c.customer_id = cp.customer_id
)
SELECT
  cs.customer_id,
  cs.first_name,
  cs.last_name,
  cs.total_rental,
  cs.last_rental_date,
  cs.total_payments,
  cs.customer_segment
FROM customer_segment cs
ORDER BY cs.total_payments DESC,
cs.total_rental DESC;
"""
# Replace 'your_database.db' with the actual path to your database
df_high_value_customers = fetch_query(query, db_path='/content/drive/MyDrive/Capstone project/sakilla database.db')
print(df_high_value_customers)
```

```python
# visuals for customer segment based on their rental behaviour
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

# Convert 'last_rental_date' to datetime
df_high_value_customers['last_rental_date'] = pd.to_datetime(df_high_value_customers['last_rental_date'])

# Calculate days since last rental (assuming 'last_rental_date' is in datetime format)
reference_date = datetime.datetime.now()
df_high_value_customers['days_last_rental'] = (reference_date - df_high_value_customers['last_rental_date'])
df_high_value_customers['days_since_last_rental'] = (reference_date - df_high_value_customers['last_rental_date']).dt.days

# Create scatter plot with color intensity representing time
sns.scatterplot(x='total_rental', y='total_payments', hue='customer_segment',
                size='days_since_last_rental', sizes=(20, 200),
                palette='viridis', data=df_high_value_customers)
plt.xlabel('Total Rentals')
plt.ylabel('Total Payments')
plt.title('Total Rentals vs. Total Payments by Customer Segment (Colored by Recency)')
plt.show()
```

```python
#Revenue Analysis
query = """
SELECT
  cat.name AS category,
  f.rating,
  c.first_name || ' ' || c.last_name AS customer_name,
  SUM(p.amount) AS revenue
FROM payment p
JOIN rental r ON p.rental_id = r.rental_id
JOIN inventory i ON r.inventory_id = i.inventory_id
JOIN film f ON i.film_id = f.film_id
JOIN film_category fc ON f.film_id
JOIN category cat ON fc.category_id = cat.category_id
JOIN customer c ON r.customer_id = c.customer_id
GROUP BY
  cat.name,
  f.rating,
  customer_name
ORDER BY
  revenue DESC;
"""
# Replace 'your_database.db' with the actual path to your database
revenue_by_category = fetch_query(query, db_path='/content/drive/MyDrive/Capstone project/sakilla database.db')
print(revenue_by_category)
```

```python
#inventory management
query = """
WITH rental_counts AS (
  SELECT
    f.title AS title,
    COUNT(r.rental_id) AS rental_count
  FROM inventory i
  LEFT JOIN rental r ON i.inventory_id = r.inventory_id
  LEFT JOIN film f ON i.film_id = f.film_id
  GROUP BY i.film_id
),
average_rental_counts AS (
  SELECT
    AVG(rental_count) AS avg_rental_count
  FROM rental_counts
),
inventory_turnover AS (
  SELECT
    f.film_id,
    f.title,
    COUNT(i.inventory_id) AS total_inventory,
    COALESCE(rc.rental_count, 0) AS rental_count,
    (COALESCE(rc.rental_count, 0) - arc.avg_rental_count) AS rental_diff
  FROM film f
  LEFT JOIN inventory i ON f.film_id = i.film_id
  LEFT JOIN rental_counts rc ON f.title = rc.title
  CROSS JOIN average_rental_counts arc
  GROUP BY f.film_id, f.title
)
SELECT
  il.film_id,
  il.title,
  il.total_inventory,
  il.rental_count,
  il.rental_diff,
  CASE
    WHEN il.rental_diff > 0 THEN 'Consider increasing inventory'
    WHEN il.rental_diff < 0 THEN 'Consider decreasing inventory'
    ELSE 'Inventory level is optimal'
  END AS recommendation,
  r.rental_id,
  i.inventory_id,
  i.store_id
FROM inventory_turnover il
LEFT JOIN inventory i ON il.film_id = i.film_id
LEFT JOIN rental r ON i.inventory_id = r.inventory_id
GROUP BY il.film_id, il.title, il.total_inventory, il.rental_count, il.rental_diff, r.rental_id, i.inventory_id, i.store_id
ORDER BY il.rental_diff DESC;
"""
# Replace 'your_database.db' with the actual path to your database
inventory_optimization_management = fetch_query(query, db_path='/content/drive/MyDrive/Capstone project/sakilla database.db')
print(inventory_optimization_management)
```

```python
sns.histplot(data=inventory_optimization_management, x='rental_diff', hue='recommendation',
             multiple='stack', kde=True)
plt.xlabel('Rental Difference')
plt.ylabel('Count')
plt.title('Distribution of Rental Difference by Recommendation')
plt.show()
```

```python
#churn analysis
query = """
SELECT
  c.customer_id,
  c.first_name,
  c.last_name,
  c.email,
  MAX(r.rental_date) AS last_rental_date,
  COUNT(r.rental_id) AS total_rentals,
  AVG(p.amount) AS average_rental_amount,
  (
    SELECT
      cat.name
    FROM category cat
    JOIN film_category fc ON cat.category_id = fc.category_id
    JOIN film f ON fc.film_id = f.film_id
    JOIN inventory i ON f.film_id = i.film_id
    JOIN rental r2 ON i.inventory_id = r2.inventory_id
    WHERE
      r2.customer_id = c.customer_id
    GROUP BY
      cat.name
    ORDER BY
      COUNT(r2.rental_id) DESC
    LIMIT 1
  ) AS most_frequent_category
FROM customer c
LEFT JOIN rental r ON c.customer_id = r.customer_id
LEFT JOIN payment p ON r.rental_id = p.rental_id
GROUP BY
  c.customer_id,
  c.first_name,
  c.last_name,
  c.email
HAVING
  last_rental_date < DATE('now', '-6 months')
ORDER BY
  last_rental_date DESC
LIMIT 10;
"""
# Replace 'your_database.db' with the actual path to your database
inactive_customers = fetch_query(query, db_path='/content/drive/MyDrive/Capstone project/sakilla database.db')
print(inactive_customers)
```

```python
#Rental Duration Analysis
query = """
SELECT
  r.rental_id,
  r.rental_date,
  julianday(r.return_date) - julianday(r.rental_date) AS rental_duration,
  c.customer_id,
  c.first_name,
  c.last_name,
  p.amount,
  f.rental_rate
FROM rental r
JOIN customer c ON r.customer_id = c.customer_id
LEFT JOIN payment p ON r.rental_id = p.rental_id
JOIN inventory i ON r.inventory_id = i.inventory_id
JOIN film f ON i.film_id = f.film_id
WHERE r.return_date IS NOT NULL
ORDER BY rental_duration DESC;
  """
# Replace 'your_database.db' with the actual path to your database
rental_duration = fetch_query(query, db_path='/content/drive/MyDrive/Capstone project/sakilla database.db')
print(rental_duration)

# Additional analysis for trends and anomalies
avg_rental_duration = rental_duration['rental_duration'].mean()

print(f"Average rental duration:", avg_rental_duration)
```
